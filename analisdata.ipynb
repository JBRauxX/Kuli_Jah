```
import numpy as np
from pylab import *
import pandas as pd
data = pd.read_csv('drive/My Drive/suaraburung/datum2.csv')
#data2 = data2.drop(['min'],axis=1) # hapus min 
data
```

# **Sample** **pola** **keseluruhan** **tanpa** **normalisasi**

```
from sklearn.preprocessing import LabelEncoder

data = data.drop(['filename'],axis=1) # hapus filename 
X = np.array(data.iloc[:, :-1])
CL = data.iloc[:, -1]
encoder = LabelEncoder()
y = encoder.fit_transform(CL)
```

```
%matplotlib inline

plt.figure(figsize=(8, 6))
plt.scatter(X[:,0], X[:,1], c=y.astype(np.float))
plt.show()
```

**Data** **ternormalisasi** 

```
from sklearn.preprocessing import LabelEncoder, StandardScaler

dt = pd.read_csv('drive/My Drive/suaraburung/datum2.csv')
dt = dt.drop(['filename'],axis=1) # hapus filename 
#dt

scaler = StandardScaler()
dfitur = scaler.fit_transform(np.array(dt.iloc[:, :-1], dtype = float))
#dfitur = np.array(dt.iloc[:, :-1])
CL = dt.iloc[:, -1]
encoder = LabelEncoder()
lbel = encoder.fit_transform(CL)

%matplotlib inline

plt.figure(figsize=(8, 6))
plt.scatter(dfitur[:,0], dfitur[:,1], c=lbel.astype(np.float))
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], c=y.astype(np.float))
```
